Linear Regression model to predict the salary based on the years of experience

```python
#importing the dependencies
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import Linear_regression as lr   # pre_existing python file with linear regression model
```

Data preprocessing

```python
#loading the data from csv file to pandas dataframe
salary_data=pd.read_csv('salary_data.csv')
```

```python
salary_data.head()
```

```python
salary_data.tail()
```

```python
salary_data.shape
```

```python
#checking missing values
salary_data.isnull().sum()
```

Splitting feature and target

```python
x=salary_data.iloc[:,:-1].values
y=salary_data.iloc[:,1].values
```

```python
print(x)
```

```python
print(y)
```

Splitting the dataset into training and tesing datas

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.33,random_state=2)  # 33% of the data will be considered for testing
```

Training the linear regression model

```python
model=lr.Linear_Regression(learning_rate=0.02,no_of_iterations=1000)
```

```python
model.fit(x_train,y_train) #training the model with x_train and y_train values
```

```python
#printing the weights and bias
print("weights: ",model.w[0])
print('bias: ',model.b)
```

Predicting the salary value for the test data

```python
test_data_prediction=model.predict(x_test)
```

```python
print(test_data_prediction)
```

Visualisation of the predicted values and actual values

```python
plt.scatter(x_test,y_test,color='red')
plt.plot(x_test,test_data_prediction,color='blue')
plt.xlabel('Work Experience')
plt.ylabel('Salary')
plt.title('Salary vs Experience')
plt.savefig("output.png") 
```
